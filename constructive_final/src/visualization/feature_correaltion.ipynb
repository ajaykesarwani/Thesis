# Stylistic Features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("./data/processed/C3_anonymized_fnd_n_s.csv")

# Apply the constructiveness threshold
df['constructive'] = df['constructive'].apply(lambda x: 1.0 if x > 0.5 else 0.0)

# Select specific columns
selected_columns = ["constructive", "Quotes", "Punc_Freq", "Punc_Type", "Exclamation", "Stopwords", "Camelcase", 
                    "Negation", "Propn", "Usermention", "Hashtags", "Misspel", "OutofVocab", "Noun", "Past_Tense", 
                    "Verbs", "Interrogation"]
df = df[selected_columns]

# Compute correlation and sort by correlation with "constructive" column
corr = df.corr()
corr_with_constructive = corr[['constructive']].drop('constructive')

# Calculate the mean percentages for each feature
mean_values = df.groupby('constructive').mean().transpose()
percentage_values = mean_values.div(mean_values.sum(axis=1), axis=0) * 100

# Plot the combined graph
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot bar plot for mean percentages
percentage_values.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'], ax=ax1)
ax1.set_title('Mean Percentage of Stylistic Features and Correlation with Constructiveness')
ax1.set_xlabel('Stylistic Features')
ax1.set_ylabel('Percentage')
ax1.set_xticklabels(percentage_values.index, rotation=90)
ax1.legend(title='Constructiveness', labels=['Non-constructive', 'Constructive'])

# Create a secondary y-axis for correlation
ax2 = ax1.twinx()
corr_with_constructive.plot(kind='line', marker='o', color='black', ax=ax2)
ax2.set_ylabel('Correlation with Constructiveness')
ax2.grid(False)
ax2.set_ylim(-1, 1)

# Adjust layout
fig.tight_layout()

# Print correlation table
print("Correlation with Constructiveness:")
print(corr_with_constructive)

# Print Non-constructive Mean and Constructive Mean
print("\nNon-con Mean:")
print(mean_values[0])
print("\nCon Mean:")
print(mean_values[1])
print("\nNon-constructive Mean %:")
print(percentage_values[0])
print("\nCon Mean %:")
print(percentage_values[1])
plt.show()
```

# POS TAG as per the Penn TreeBank

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Function to process text, find common POS tags, and calculate occurrences per sentence
def process_text(text):
    # Tokenize the text into sentences
    sentences = sent_tokenize(text)
    # Initialize dictionary to store POS tag occurrences per sentence
    pos_counts = {}
    # Count the number of sentences
    num_sentences = len(sentences)
    # Common POS tags
    common_pos_tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS',
                       'LS', 'MD', 'NN', 'NNS', 'NNP', 'NNPS', 'PDT', 'POS',
                       'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO',
                       'UH', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT',
                       'WP', 'WP$', 'WRB']
    # Process each sentence
    for sentence in sentences:
        # Tokenize the sentence into words and get POS tags
        words = word_tokenize(sentence)
        pos_tags = pos_tag(words)
        # Extract POS tags from tuples and count occurrences for common tags
        for word, tag in pos_tags:
            if tag in common_pos_tags:
                pos_counts[tag] = pos_counts.get(tag, 0) + 1
    # Divide the count of each POS tag by the number of sentences
    for key in pos_counts:
        pos_counts[key] /= num_sentences
    return pos_counts

# Read text from CSV file
df = pd.read_csv("./data/processed/C3_anonymized.csv", usecols=["pp_comment_text", "constructive"])

# Apply POS tagging and count POS tag occurrences for each sentence
df["POS_counts"] = df["pp_comment_text"].apply(process_text)

# Extract common POS tags as columns and their counts as values
common_pos_tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS',
                   'LS', 'MD', 'NN', 'NNS', 'NNP', 'NNPS', 'PDT', 'POS',
                   'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO',
                   'UH', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT',
                   'WP', 'WP$', 'WRB']
for pos_tag in common_pos_tags:
    df[pos_tag] = df["POS_counts"].apply(lambda x: x.get(pos_tag, 0))

# Drop the original "POS_counts" column
df.drop(columns=["POS_counts"], inplace=True)

# Write the results to a new CSV file
df.to_csv("./data/processed/pos_tag_nltk_common.csv", index=False)
```
